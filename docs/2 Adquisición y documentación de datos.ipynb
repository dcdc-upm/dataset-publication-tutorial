# Adquisición y documentación de los datos

Para este tutorial vamos a obtener el dataset [Accidentes de tráfico de la ciudad de Madrid](https://datos.madrid.es/portal/site/egob/menuitem.c05c1f754a33a9fbe4b2e4b284f1a5a0/?vgnextoid=7c2843010d9c3610VgnVCM2000001f4a900aRCRD&vgnextchannel=374512b9ace9f310VgnVCM100000171f5a0aRCRD&vgnextfmt=default) publicado por el Ayuntamiento de Madrid.

1. Lo primero que hay que hacer cuando se va a utilizar un dataset existente es mirar la licencia de uso. En este caso, para los datos de la ciudad de Madrid la licencia está publicada [AQUÍ](https://datos.madrid.es/egob/catalogo/aviso-legal). Los puntos en los que nos tenemos que fijar cuando analizamos una licencia son los siguientes:

    - **¿Se pueden reutilizar los datos?**

        *'Las presentes condiciones generales **permiten** la reutilización de los documentos y datos sometidos a ellas para fines comerciales y no comerciales'*.

    - **¿Bajo que condiciones se pueden reutilizar?**

        *'Debe citarse la fuente de los documentos objeto de la reutilización [...] prohibido desnaturalizar el sentido de la información [...] prohibido realizar labores de re-identificación de personas [...]'*

    - **¿Puedo publicar los datos que he reutilizado y transformado?**

        *'Esta autorización conlleva, asimismo, la cesión gratuita y no exclusiva de los derechos de propiedad intelectual, en su caso, correspondientes a tales documentos, **autorizándose** la realización de actividades de reproducción, **distribución**, comunicación pública **o transformación**, necesarias para desarrollar la actividad de reutilización autorizada, en cualquier modalidad y bajo cualquier formato, para todo el mundo y por el plazo máximo permitido por la Ley.'* 


2. Una vez sabemos que podemos hacer uso y publicación del dataset, lo descargamos y empezamos a documnetar el proceso para ser transparente.
    
    *En este trabajo toda esta documentación la haremos en el `README.md` como es común en este tipo de trabajos.*

3. A continuación realizaremos las tareas de transformación y aplicaremos los filtros que queramos para conseguir el resultado esperado. 

    *Al igual con el paso anterior esto se irá documentando en el `README.md` mientras se realiza.*

4. Cuando tengamos el reurso preparado para su publicación lo revisaremos manualmente y crearemos un datasheet especial para la publicación en un repositorio digital.
    
    *El datasheet debe incluir tanta información útil como sea necesaria para la reutilización del recurso como para que un usuario interesado pueda reconocer fácilmente si el recurso le interesa o no. Recomendamos prestar especial atención a artículos como el siguiente para construir un datasheet de calidad:*
    
    *Gebru, Timnit, Jamie Morgenstern, Briana Vecchione, et al. “Datasheets for Datasets.” Commun. ACM 64, no. 12 (2021): 86–92. https://doi.org/10.1145/3458723.*


5. Publicaremos el recurso y nos aseguraremos de que todo está tal y como queremos.
